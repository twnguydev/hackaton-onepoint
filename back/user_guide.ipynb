```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Charger le fichier NetCDF
fichier = "../back/Dataset/E-OBS Copernicus/tx_ens_mean_0.1deg_reg_2011-2022_v26.0e.nc"
dataset = nc.Dataset(fichier, 'r')
variable = "tx"

# Extraire les latitudes et longitudes
lats = dataset.variables['latitude'][:]
lons = dataset.variables['longitude'][:]

# Dfinir les coordonnes spcifiques
latitude_cible = 43.30885  # Exemple de latitude
longitude_cible = 5.36788  # Exemple de longitude

# Fonction pour trouver l'indice le plus proche d'une valeur dans un tableau
def trouver_indice_plus_proche(tableau, valeur):
    indice = np.abs(tableau - valeur).argmin()
    return indice

# Trouver les indices des lat/lon les plus proches
indice_lat = trouver_indice_plus_proche(lats, latitude_cible)
indice_lon = trouver_indice_plus_proche(lons, longitude_cible)

# Extraire la variable  un moment donn
time_index = 0  # Choisir l'indice temporel souhait
var_at_time = dataset.variables[variable][time_index, :, :]  # Donnes pour l'instant t

# Extraire la valeur  la lat/lon spcifie
valeur_specifique = var_at_time[indice_lat, indice_lon]
print(f"Valeur de {variable}  la latitude {latitude_cible} et longitude {longitude_cible}: {valeur_specifique}")

# Visualisation des donnes
plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
contour = plt.contourf(lons, lats, var_at_time, cmap='Blues', transform=ccrs.PlateCarree())
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, zorder=0, edgecolor='black')
ax.add_feature(cfeature.OCEAN)

# Afficher la barre de couleur
plt.colorbar(contour, label=f'{variable} (C)')
plt.show()

# Fermer le fichier NetCDF
dataset.close()
```

```python
import netCDF4 as nc
import numpy as np
import json
from datetime import datetime, timedelta
import warnings
import xarray as xr
import pandas as pd

warnings.filterwarnings("ignore", category=UserWarning, message="WARNING: missing_value not used")

def ms_to_kmh(speed_ms):
    return speed_ms * 3.6

def lire_netcdf_et_creer_json(fichier_nc, variable, lat, lon, date_debut, date_fin):
    try:
        ds = xr.open_dataset(fichier_nc)

        donnees = ds[variable].sel(latitude=lat, longitude=lon, method='nearest')

        dates = donnees.time.to_index()

        date_debut = pd.to_datetime(date_debut)
        date_fin = pd.to_datetime(date_fin)

        # Slectionner les donnes entre date_debut et date_fin
        donnees_periode = donnees.sel(time=slice(date_debut, date_fin))

        resultats = {
            "metadata": {
                "variable": variable,
                "latitude": float(donnees.latitude.values),
                "longitude": float(donnees.longitude.values),
                "unite": "km/h"  # Chang de m/s  km/h
            },
            "donnees": []
        }

        for date, valeur in zip(donnees_periode.time.values, donnees_periode.values):
            valeur_kmh = ms_to_kmh(float(valeur)) if not np.isnan(valeur) else None
            resultats["donnees"].append({
                "date": pd.Timestamp(date).strftime("%Y-%m-%d"),
                "valeur": round(valeur_kmh, 2) if valeur_kmh is not None else None
            })

        return resultats

    except Exception as e:
        print(f"Une erreur s'est produite : {str(e)}")
        return None

    finally:
        if 'ds' in locals():
            ds.close()

def main():
    fichier_nc = "../back/Dataset/E-OBS Copernicus/fg_ens_mean_0.1deg_reg_2011-2022_v26.0e.nc"
    variable = "fg"
    latitude = 43.30885
    longitude = 5.36788
    date_debut = "2020-01-01"
    date_fin = "2022-12-31"  # Ajustez cette date si ncessaire

    resultats = lire_netcdf_et_creer_json(fichier_nc, variable, latitude, longitude, date_debut, date_fin)

    if resultats:
        with open('resultats_vent_2020_2022_marseille_kmh.json', 'w') as f:
            json.dump(resultats, f, indent=2)
        print("Les donnes ont t extraites et sauvegardes dans 'resultats_vent_2020_2022_marseille_kmh.json'")
    else:
        print("chec de l'extraction des donnes.")

if __name__ == "__main__":
    main()
```
